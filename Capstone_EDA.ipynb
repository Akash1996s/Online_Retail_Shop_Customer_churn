```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = 'whitegrid')
```

```python
data = pd.read_excel('Store_Data_after_EDA.xlsx')
```

```python
data.columns
```

```python
data = data.drop(columns=['Unnamed: 0', 'index'],axis=1)
```

```python
data.head()
```

```python
data.info()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.boxplot(x = data.esent,color='red');
plt.subplot(2,2,2)
sns.boxplot(x = data.eclickrate,color='green')
plt.subplot(2,2,3)
sns.boxplot(x = data.avgorder,color='pink')
plt.subplot(2,2,4)
sns.boxplot(x = data.ordfreq,color='yellow')
plt.show()
```
